```python
import matplotlib.pyplot as plt
import numpy as np

# --- Updated Y Labels ---
# Flipped from 1:X to X:1
sizes = [25, 10, 5, 2, 1]
y_labels = [f'{size}:1' for size in sizes]

# --- Data Extraction Helper ---
def extract_values(input_lists, n):
    data = {}
    for method_list in input_lists:
        if not method_list:
            continue
        method_name = method_list[0][0]
        data[method_name] = [tup[2][n] for tup in method_list]
    return data

def plot_heatmap(ax, Z, title, xticks, yticks, cmap='plasma', is_top_row=True, is_left_col=True):
    # aspect='equal' keeps cells square
    im = ax.imshow(Z, cmap=cmap, origin='lower', aspect='equal')
    
    # X-axis: Always show the names (ADASYN, etc.)
    ax.set_xticks(np.arange(len(xticks)))
    ax.set_xticklabels(xticks, rotation=45, ha='right', fontsize=7)
    
    # Y-axis: Always show the flipped ratios (25:1, etc.) on the left column
    ax.set_yticks(np.arange(len(yticks)))
    ax.set_yticklabels(yticks if is_left_col else [], fontsize=7)
    
    # Axis Titles: Only show labels for the top row
    if is_top_row:
        if is_left_col:
            ax.set_ylabel('Class Imbalance Ratio', fontsize=8)
        ax.set_xlabel('Oversampling Method', fontsize=8)
    
    ax.set_title(title, fontsize=9)
    
    # Annotate cells
    for i in range(Z.shape[0]):
        for j in range(Z.shape[1]):
            ax.text(j, i, f'{Z[i,j]:.2f}', ha='center', va='center',
                    fontsize=6, color='white' if Z[i,j] < Z.max()/2 else 'black')
    return im

# -----------------------------
# Create figure: 4 rows × 2 columns
# -----------------------------
fig, axes = plt.subplots(4, 2, figsize=(7, 14))

for row_idx, (name, data) in enumerate(all_search_data):
    data_recall = extract_values(data, n=1)
    data_runtime = extract_values(data, n=2)
    labels = list(data_recall.keys())
    
    Z_recall = np.array([data_recall[label] for label in labels]).T
    Z_runtime = np.array([data_runtime[label] for label in labels]).T

    is_top = (row_idx == 0)

    # Column 0: Recall
    plot_heatmap(axes[row_idx, 0], Z_recall, f'{name} - Recall', 
                 labels, y_labels, cmap='plasma', 
                 is_top_row=is_top, is_left_col=True)
    
    # Column 1: Runtime
    plot_heatmap(axes[row_idx, 1], Z_runtime, f'{name} - Runtime', 
                 labels, y_labels, cmap='viridis_r', 
                 is_top_row=is_top, is_left_col=False)

# -----------------------------
# Final Spacing Adjustments
# -----------------------------
plt.subplots_adjust(
    left=0.15, 
    right=0.95, 
    top=0.95, 
    bottom=0.05, 
    wspace=0.0, 
    hspace=0.5
)

plt.show()
```
